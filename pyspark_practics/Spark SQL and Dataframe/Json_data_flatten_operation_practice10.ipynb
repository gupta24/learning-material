------------- Read Json Data

```
# read the json data from file path..
json_df = spark.read.format('json')\
                    .option('header', 'true')\
                    .option('inferschema', 'true')\
                    .option('multiline', 'true')\
                    .load('/FileStore/tables/resturant_json_data.json')
json_df.display()
```

```
json_df.printSchema()
```

--------------- Json Data Flatten

```
# import libraries..
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
# flatten the json data where explode and explode_outer parameter are used
# when we want to convert the array data into struct for easy access.
# here different b/w explode (use when column data not having any null value) and explode_outer (use when column data having null value)
# explode skip entire dataframe data when get any null field in column while explode_outer is not do this thing...

json_df.select('*', explode('restaurants').alias('new_restaurants')).printSchema()
```

```
json_df1 = json_df.select('*', explode('restaurants').alias('new_restaurants')).drop('restaurants')\
                 .select('*',
                          'new_restaurants.restaurant.R.res_id',
                          explode_outer('new_restaurants.restaurant.establishment_types').alias('restaurant_estiblished'),
                          'new_restaurants.restaurant.cuisines',
                          'new_restaurants.restaurant.location.address',
                          'new_restaurants.restaurant.location.city',
                          'new_restaurants.restaurant.name'
                        ).select('*',
                           explode_outer('new_restaurants.restaurant.offers').alias('cuisines_offers')
                        )

json_df1 = json_df1.drop('new_restaurants')                    
json_df1.printSchema()
json_df1.drop('code','message','results_found','results_shown','results_start','status').distinct().display()
                
```
